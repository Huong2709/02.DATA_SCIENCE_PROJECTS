```python
# create function to extract emojicons
def extract_emojis(doc):
  lst_emojis = []
  for i in doc:
    if i in emoji.UNICODE_EMOJI['en']: lst_emojis.append(i)
  return lst_emojis
```

#### Text cleaning
Convert emojicons to words: 
* After extract emojicons in all documents, scan each emojicon and define the meaning words.
* Load edited emojicons for data cleaning.

Word_tokenize for Vietnamese documents:
* Convert to lower words.
* Use Underthesea library to convert to word_tokenize. Ex: "sạch sẽ" => "sạch_sẽ".

Remove some special symbols:
* List down list of special symbols, such as: '~','`','!','@','#','$',...
* Remove these symbols.
* Remove some typing mistakes: "  " (2 spaces), " _" (1 space, 1 underscore), "_ " (1 underscore, 1 space)

Link some special words with other words:
* Link "không", "ko", "kg", "chả", "chẳng" to the word right behind. Ex: "không thích" => "không_thích"

Remove stopwords:
* First, do all above text cleaning steps and use CountVectorize to convert to bag of words.
* Save this bag of words to csv file.
* Scan each words and define what words will be removed, what words will be remained.
* Save the chosen words to txt file.
* Load chosen words, if words in document are in chosen words, they will be remaned. If not, they will be removed.

```python
# load edited emojicon
file = open('lib/files/emojicon.txt', 'r', encoding="utf8")
emoji_lst = file.read().split('\n')
file.close()

emoji_dict = {}
lst_key = []
lst_value = []
for line in emoji_lst:
  key, value = line.split('\t')
  lst_key.append(key)
  lst_value.append(value)
  emoji_dict[key] = str(value)
```

```python
# load chosen words
file = open('lib/files/chosen_words_full.txt', 'r', encoding="utf8")
lst_chosen_words = file.read().split('\n')
file.close()
```

```python
lst_spec_symbol = ['~','`','!','@','#','$','%','^','&','*','(',')','-','=','+',   '[','{','}','}','\\','|',';',':','"',     ',','<','.','>','/','?']
```

```python
lst_find_words = ['không ','ko ','kg ','chẳng ','chả ']
lst_replace_words = ['không_','ko_','kg_','chẳng_','chả_']
```

```python
start_time = time.time()

lst_document = []
for doc in document:
  doc = doc.lower()

# replace emojicons
  for i in range(len(lst_key)):
    doc = doc.replace(lst_key[i], ' '+lst_value[i])
  for j in doc:
    if j in emoji.UNICODE_EMOJI['en']: 
      doc = doc.replace(j, '')

# word tokenize: sạch sẽ => sạch_sẽ
  doc = word_tokenize(doc, format='text')

# remove special symbols
  rx = '[' + re.escape(''.join(lst_spec_symbol)) + ']' 
  doc = re.sub(rx, '', doc)

  doc = doc.replace('  ',' ')
  doc = doc.replace(' _',' ')
  doc = doc.replace('_ ',' ')

# replace 'không ' to 'không_' co link words, v.v...
  for i in range(len(lst_find_words)):
    doc = doc.replace(lst_find_words[i], lst_replace_words[i]) 

# remove stop_words
  lst_words = []
  for j in doc.split(' '):
    if j in lst_chosen_words: lst_words.append(j)
  doc = ' '.join(lst_words)  

  lst_document.append(doc)  
    
end_time = time.time()
```

### Choose k

```python
df_cluster = df_clean_cv.iloc[:,7:]
```

```python
df_cluster.head()
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics import silhouette_score
scores = []
for k in range(2,8):
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_cluster)
    scores.append(silhouette_score(df_cluster,kmeanModel.labels_))
print(scores)
```

```python
# suitable k
id_max = np.argmax(scores)
print('k=', id_max+2)
```

```python
# k means determine k using elbow method
distortions = []
K=range(1,8)
for k in (1,2):
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_cluster)
    print(cdist(df_cluster, kmeanModel.cluster_centers_,'euclidean'))
#     distortions.append(sum(np.min(cdist(df, kmeanModel.cluster_centers_,
#                                        'euclidean'), axis=1)) / df.shape[0])
```

```python
# k means determine k using elbow method
distortions = []
K=range(1,8)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_cluster)
    distortions.append(sum(np.min(cdist(df_cluster, kmeanModel.cluster_centers_,
                                       'euclidean'), axis=1)) / df_cluster.shape[0])
```

```python
# plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
